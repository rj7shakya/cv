```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import graphviz
```

```python
X = np.array([1,2,3,4,100,200,105,300])
y = np.array([1,1,1,1,0,0,0,0])
```

```python
X.shape
y.shape
```

```python
X= X.reshape(-1,1)
Y= y.reshape(-1,1)
```

```python
plt.plot(X,y)
```

```python
clf = DecisionTreeClassifier()
clf.fit(X,y)
```

```python
clf.predict([[4]])
```

```python
from sklearn.tree import export_graphviz
from sklearn import tree
dot_data = tree.export_graphviz(clf)
graph = graphviz.Source(dot_data,"test.pdf")
# graph.render("simple")
```
